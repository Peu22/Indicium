```python
# Temos dois campos importantes para analisar: o campo "disponibilidade_365", que indica quantos dias o anúncio está disponível para reserva ao longo do ano,
# e o campo "numero_de_reviews", que mostra quantos comentários foram feitos para cada listagem.
# Se a disponibilidade é alta, isso sugere que há pouca procura na região. Por outro lado, se há muitos comentários, isso indica que o local é bastante procurado.
# Portanto, queremos identificar o primeiro bairro em que ocorre uma combinação de baixa disponibilidade e alto número de comentários, o que sugere que há uma alta demanda apesar da baixa disponibilidade.

# Primeiro, calculamos a disponibilidade média de aluguel ao longo do ano para cada bairro
disponibilidade_media = data.groupby('bairro')['disponibilidade_365'].mean()

# Ordenamos os bairros com base na disponibilidade média (do menor para o maior)
bairros_por_disponibilidade = disponibilidade_media.sort_values().index.tolist()

# Calculamos o número total de reviews para cada bairro
total_reviews = data.groupby('bairro')['numero_de_reviews'].sum()

# Ordenamos os bairros com base no número total de reviews (do maior para o menor)
bairros_por_reviews = total_reviews.sort_values(ascending=False).index.tolist()

# Encontramos a primeira área que ocorre em ambas as listas
primeiro_bairro_com_ocorrencia_comum = next(bairro for bairro in bairros_por_disponibilidade if bairro in bairros_por_reviews)

# Encontramos o bairro correspondente à primeira área encontrada
bairro_group_correspondente = data[data['bairro'] == primeiro_bairro_com_ocorrencia_comum]['bairro_group'].iloc[0]


print(f"A área mais indicada para compra é {primeiro_bairro_com_ocorrencia_comum}, no bairro {bairro_group_correspondente}" )
```

# ***Questão 3:***

As principais variáveis que podemos considerar para prever o preço incluem [longitude], [latitude], por causa que dão a localização em números e [disponibilidade_365], [calculado_host_listings_count], [reviews_por_mes], pois segundo o mapa de calor, são as variáveis de maior correlação com o preço

Como desejamos prever o preço (valor númerico), logo estamos resolvendo um problema de regressão

Quando estamos tentando prever os preços, usando múltiplas características que podem não ter uma relação linear clara com o preço, a regressão polinomial pode ser uma escolha útil. Ela nos permite capturar relações não lineares entre as características e o preço. No entanto, precisamos ter cuidado, pois a regressão polinomial tem uma tendência maior a superajustar os dados (overfitting), o que significa que o modelo pode se adaptar muito bem aos dados de treinamento, mas não generalizar bem para novos dados. Além disso, a interpretação dos resultados da regressão polinomial pode ser mais complicada do que em modelos lineares simples, o que pode dificultar a compreensão das relações entre as características e o preço.

Ao avaliar o desempenho dos modelos de regressão, várias medidas podem ser utilizadas, dependendo dos objetivos específicos do projeto. Por exemplo, se o objetivo principal é minimizar a magnitude dos erros cometidos pelas previsões, o Erro Absoluto Médio (MAE) pode ser uma métrica adequada. Por outro lado, se o objetivo é compreender o quanto o modelo explica a variabilidade presente nos dados, o Coeficiente de Determinação (R²) pode ser mais apropriado. Um valor de R² mais próximo de 1 indica que o modelo se ajusta bem aos dados, enquanto um valor mais próximo de 0 sugere um ajuste menos satisfatório.

# ***Questão 4:***

Mesmo que o "novo" apartamento ja estivesse na base de dados um dos jeitos mais simples de fazer essa predição seria da seguinte maneira:

```python
# Importando as bibliotecas
# Usando o SkLearn para importar diretamente a regreção linear
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Usando apenas as colunas que possuem números, ja que a latitude e a longitude ja me dão o local
X = data[['latitude', 'longitude', 'minimo_noites','numero_de_reviews', 'reviews_por_mes', 'calculado_host_listings_count','disponibilidade_365']]
y = data['price']
```

```python
# Modelo de Regressão linear que foi importado pelo SkLearn
modelo = LinearRegression()
```

```python
modelo.fit(X, y)
```

```python
 # Adicionando as características do novo apartamento
novo_apartamento = np.array([[40.75362, -73.98377, 1, 45, 0.38, 2, 355]])
```

```python
# Fazendo a predição do preço
preco_predito = modelo.predict(novo_apartamento)
```

```python
print("O preço predito do apartamento é:", preco_predito[0])
```
